```python
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.ticker import FuncFormatter

data = {
    "Date": [
        "2025-03-19", "2025-03-20", "2025-03-21", "2025-03-22", "2025-03-25",
        "2025-03-26", "2025-03-27", "2025-03-28", "2025-03-29", "2025-04-01",
        "2025-04-02", "2025-04-03", "2025-04-04", "2025-04-05", "2025-04-08",
        "2025-04-09", "2025-04-10", "2025-04-11", "2025-04-12"
    ],
    "Account Value": [
        1_000_000, 1_000_000, 1_000_000, 998_840, 999_546, 1_000_330, 996_689.15,
        996_591.35, 993_560.25, 992_074.25, 992_661.35, 993_954.95, 989_881.95,
        1_002_397.25, 1_022_073.85, 1_134_016.15, 946_793.15, 875_256.65, 829_256.85
    ]
}

df = pd.DataFrame(data)
df["Date"] = pd.to_datetime(df["Date"])

plt.figure(figsize=(10, 6))
plt.plot(df["Date"], df["Account Value"], marker='o', linewidth=2, label="Account Value")
plt.title("Portfolio Account Value Over Time", fontsize=14)
plt.xlabel("Date")
plt.ylabel("Account Value ($)")
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.gca().yaxis.set_major_formatter(FuncFormatter(lambda x, _: f'${x:,.0f}'))
plt.legend()
plt.show()
```

```python

strategies = ['Long Call Options', 'Long Put Options', 'Protective Puts', 'Straddles', 'Strangles']
trade_counts = [3, 5, 3, 3, 3]

plt.figure(figsize=(8, 6))
plt.pie(trade_counts, labels=strategies, autopct='%1.1f%%', startangle=140, colors=['#66b3ff','#99ff99','#ffcc99','#ffb3e6','#ff6666'])
plt.title('Strategy Distribution of Trades')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt


category_pl = {
    "Protective Puts": 331,
    "Straddles": -50295,
    "Long Puts": -40902,
    "Long Calls": 7976,
    "Strangles": -46970
}


category_win_pct = {
    "Protective Puts": "2/3",
    "Straddles": "1/3",
    "Long Puts": "1/5",
    "Long Calls": "1/3",
    "Strangles": "1/3"
}

abs_category_pl = {k: abs(v) for k, v in category_pl.items()}


labels = [
    f"{cat}\nP/L: ${category_pl[cat]:+,.0f}\nWin: {category_win_pct[cat]}"
    for cat in category_pl
]

fig, ax = plt.subplots(figsize=(9, 9))
colors = ['#FF9999', '#66B3FF', '#99FF99', '#FFCC99', '#C2C2F0']
explode = [0.05] * len(category_pl)

ax.pie(
    abs_category_pl.values(),
    labels=labels,
    autopct='%1.1f%%',
    startangle=140,
    colors=colors,
    explode=explode,
    shadow=True,
    textprops={'fontsize': 12}
)

ax.set_title("Profit & Loss by Strategy Category\n(With % of Winning Trades)", fontsize=16)
plt.tight_layout()
plt.show()
```
